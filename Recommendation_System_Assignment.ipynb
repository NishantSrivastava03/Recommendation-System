# EDA

#Removed_Unuseful_Columns

```python
book1=book.drop(columns='Unnamed: 0')
book1
```

```python
book2=book1.rename(columns={'User.ID':'userId','Book.Title':'bookT','book.Rating':'bookR'})
```

```python
book2
```

#number of unique users in the dataset

```python
len(book2.userId.unique())
```

```python
len(book2.bookT.unique())
```

#Converting long data into wide data using pivot table.

```python
user_book_df=book2.pivot_table(index='userId',columns='bookT').reset_index(drop=True)
user_book_df
```

```python
user_book_df.index=book2.userId.unique()
user_book_df
```

#impute those Nans with 0 values

```python
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

```python
#Calculating cosine similarity between users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim=1 - pairwise_distances(user_book_df.values,metric='cosine')
```

```python
user_sim
```

#Storing the result in dataframe

```python
user_sim_df=pd.DataFrame(user_sim)
user_sim_df
```

#Set the index and column names to user ids

```python
user_sim_df.index=book2.userId.unique()
user_sim_df.columns=book2.userId.unique()
```
